```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
df = pd.read_csv("chocolate.csv")
df.columns=["Numero de referencia","Compa√±ia","Localizacion", "Fecha","Pais","Origen", "Porcentaje","Ingredientes","Caracteristicas","Rango"]
print(df)
df['Porcentaje'] = df['Porcentaje'].str.replace('%', '').astype(float)
##Forma de graficacion.
plt.bar(df['Fecha'],df['Porcentaje'])
plt.xlabel('Fecha')
plt.ylabel('Porcentaje')
plt.title('Relacion entre X e Y')
plt.tight_layout()
plt.yticks(fontsize=6)
plt.xticks(fontsize=6)
plt.show()
# Asignamos las variables.
X = df [['Fecha']] # Variable independiente
Y = df['Porcentaje']# Variable dependiente
#Creamos el modelo de aprendizaje.
modelo = LinearRegression()
modelo.fit(X,Y) # Crea la mejor linea entre esas 2 variables
#Obtenemos las predicciones.
y_pred = modelo.predict(X)
#Vizuaizacion.
plt.scatter(df['Fecha'], df['Porcentaje'], label= 'Datos reales', color = 'green')
plt.plot(df['Fecha'], y_pred, label ='Regresion lineal', color = 'red')
plt.xlabel('Fecha')
plt.ylabel('Porcentaje')
plt.title('Regresion lineal')
plt.legend()
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.tight_layout()
plt.show()

m = modelo.coef_[0]
b = modelo.intercept_
print(f"Ecuaci√≥n de la recta: y = {m:.4f}x + {b:.4f}")
```
