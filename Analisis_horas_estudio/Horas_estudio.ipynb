```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv("datos_regresion_lineal.csv")
df.columns = ['Horas','Nota']
print(df)
#Impresion de datos originales.
plt.scatter(df['Horas'],df['Nota'],color ="blue")
plt.title('Datos originales')
plt.xlabel('Horas de estudio')
plt.ylabel('Nota estudiantes')
plt.tight_layout()
plt.xticks(size = 5)
plt.yticks(size = 5)
plt.show()

#Comienzo de codigo de RL
#Asignacion de variables
X = df[['Horas']]
Y = df['Nota']
#Creamos el modelo de aprendizaje.
modelo = LinearRegression()
modelo.fit(X,Y) #El fit es para creacion de linea que pasa sobre los datos.
#Obtenemos la prediccion.
y_pred = modelo.predict(X)
#Vizualizacion de los datos.
plt.scatter(df['Horas'],df['Nota'], label = 'Datos reales', color ="blue")
plt.plot(df['Horas'],y_pred, label = 'Regresion  lineal', color = 'red' )
plt.xlabel('Horas de estudio')
plt.ylabel('Nota en base a las hora de estudio')
plt.xticks(size = 5)
plt.yticks(size = 5)
plt.legend()
plt.tight_layout()
plt.show()

a = modelo.coef_[0]
b = modelo.intercept_
print(f"La ecuacion de la recta y es igual a: {a:.4f}x + {b:.4f}")
```
